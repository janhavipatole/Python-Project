```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL='https://mamaearth.in/product-category/lipsticks'
URL
```

```python
r=requests.get(URL)
htmldata=r.content
htmldata
```

```python
soup=BeautifulSoup(htmldata,'html.parser')
soup.prettify
```

```python
title=soup.find_all(class_="title")
print(title)
```

```python
Name=[]
for i in range (0,28):
    Name.append(title[i].get_text())
print(Name)
```

```python
descriptions=soup.find_all(class_="subtitle")
print(descriptions)
```

```python
description=[]
for i in range (0,len(descriptions)):
    description.append(descriptions[i].get_text())
print(description)
```

```python
Rev=soup.find_all(class_="review-count")
print(Rev)
```

```python
Rev_count=[]
for i in range (0,len(Rev)):
    Rev_count.append(Rev[i].get_text())
print(Rev_count)
```

```python
stars=soup.find_all(class_="Flex-sc-1lsr9yp-0 cHTzIe rating-review")
print(stars)
```

```python
star=[]
Stars=[]
STR=[]
for i in range(0,len(stars)):
    star.append(stars[i].get_text())
for i in range(0,len(stars)):
    Stars.append(star[i].strip())
STR=[i[:+5] for i in Stars]
print(STR)
```

```python
price=soup.find_all(class_="price")
print(price)
```

```python
product_price=[]
for i in range(0,len(price)):
    product_price.append(price[i].get_text())
print(product_price)
```

```python
import pandas as pd 
df=pd.DataFrame({"lipsticks_name":Name,"review-count":Rev_count,"price":product_price,"product_discription":description,"Review_stars":STR},index=range(0,len(price)))
print(df)
```

```python
import os 
df.to_csv("mamaearth lipsticks.csv",index=False)
a=pd.read_csv("mamaearth lipsticks.csv")
a
```
